```python
dfNew = df.drop(columns=['Full Name', 'Korean Name', 'K. Stage Name', 'Other Group'])
dfNew.shape
```

```python
dfNew.head()
```

```python
cnt = 0
  
visited = []

for i in range(0, len(dfNew['Gender'])):
    
    if dfNew['Gender'][i] not in visited: 
        
        visited.append(dfNew['Gender'][i])
          
        cnt += 1
  
print("No.of.unique values :",
      cnt)
  
print("unique values :",
      visited)
```

```python
maleCount = 0 
femaleCount = 0 
otherCount = 0

for i in dfNew['Gender']: 
  if i == 'M': 
    maleCount += 1
  elif i == 'F': 
    femaleCount += 1 
  else: 
    otherCount += 1

genderList = [maleCount, femaleCount, otherCount]
genders = np.array(genderList)
gend = 'Male', 'Female', 'Other'
explode = (0.1, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(genders, labels=gend, explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.show() 
```

```python
dfNew['Group'].fillna('Solo', inplace = True)
dfNew['Birthplace'].fillna('Unknown', inplace = True)
```

```python
cnt = 0
  
groupVisited = []

for j in range(0, len(dfNew['Group'])):
    
    if dfNew['Group'][j] not in groupVisited: 
        
        groupVisited.append(dfNew['Group'][j])
          
        cnt += 1
  
print("No.of.unique values :",
      cnt)
  
print("unique values :",
      groupVisited)
```

```python
# trying to create a list, where each element in groupVisited has the number of idols belonging to it. 

occurenceList = []
listing = [] 

groupList = dfNew['Group'].tolist()

def countElement(a):
  g = {}
  for i in a:
    if i in g:
      g[i] +=1
    else:
      g[i] =1
  return g

groupDict = countElement(groupList)

myList = groupDict.items()
myList = sorted(myList) 
x, y = zip(*myList)

fig = plt.figure(figsize =(10, 60))
plt.xticks(rotation=90)
plt.xlabel("Number of Idols per Group")
plt.ylabel("Group Name")
plt.barh(x, y, color ='cyan')
```
